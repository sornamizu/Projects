```python
import pandas as pd
df = pd.read_csv("netflix_titles.csv")

print("Shape of dataset:", df.shape)
print(df.head())
```

```python
# Check how many missing values remain
print("Missing values after cleaning:\n", df.isnull().sum())

# Handle missing 'duration'
# Option 1: Drop rows with missing duration
df = df.dropna(subset=['duration'])

# Option 2: Fill with a placeholder (if you want to keep them)
# df['duration'] = df['duration'].fillna("Unknown")

# Verify again
print("Missing values after fixing duration:\n", df.isnull().sum())
```

```python
df['genre'] = df['listed_in'].str.split(',').str[0].str.strip()
df['rating'] = df['rating'].str.upper().str.strip()
df['country'] = df['country'].str.strip()

print("Unique genres (sample):", df['genre'].unique()[:10])
print("Unique ratings:", df['rating'].unique())
print("Unique countries (sample):", df['country'].unique()[:10])
```
